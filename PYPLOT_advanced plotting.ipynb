## 1. moving to object-based plotting

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure()
plt.subplot(1,2,1) #1행 2열 1번째
plt.subplot(1,2,2) #1행 2열 2번째
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
```

```python
fig=plt.figure()  #create a figure object
ax=fig.add_subplot(1,1,1) # create an axes object in the figure
ax.plot([1,2,3,4])
ax.set_ylabel('some numbers')
```

```python
fig1=plt.figure()
fig2=plt.figure()
ax1=fig1.add_subplot(1,1,1)
ax2=fig2.add_subplot(2,1,1)
ax3=fig2.add_subplot(2,1,2)

fig.canvas.draw()
```

## 4. twin axes

```python
import numpy as np

fig=plt.figure()
ax1=fig.add_subplot(1,1,1)
ax2=ax1.twinx()  #create a  new set of axes(ax2) that shares the x-axis with ax1

t=np.linspace(0., 10., 100)
ax1.plot(t,t**2,'b-')
ax2.plot(t, 1000 / (t + 1), 'r-')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Density (cgs)', color='red')
ax2.set_ylabel('Temperature (K)', color='blue')
```

## 5. controlling the appearance of plots

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.set_title("My plot", fontsize='large')
for tick in ax.xaxis.get_ticklabels():
    tick.set_fontsize('large')
    tick.set_fontname('Times New Roman')
    tick.set_color('blue')
    tick.set_weight('bold')

for tick in ax.yaxis.get_ticklabels():
    tick.set_fontsize('small')
    tick.set_fontname('Times New Roman')
    tick.set_color('red')
    tick.set_weight('bold')    
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
title = ax.set_title("My plot", fontsize='large')
points = ax.scatter([1,2,3], [4,5,6]) #(1,4) , (2,5), (3,6)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
title = ax.set_title("My plot", fontsize='large')
points = ax.scatter([1,2,3], [4,5,6])
points.set_color('red')
title.set_fontsize('x-large')
fig.canvas.draw()
```

## 6.  rc parameters

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.rc('xtick', color='r', labelsize='medium', direction='out')
plt.rc('xtick.major', size=4, pad=4)
plt.rc('xtick.minor', size=2, pad=4)
plt.rcdefaults() #reset the parameters to thier default values

#xtick.major.size     : 4      # major tick size in points
#xtick.minor.size     : 2      # minor tick size in points
#xtick.major.pad      : 4      # distance to major tick label in points
#xtick.minor.pad      : 4      # distance to the minor tick label in points
#xtick.color          : r      # color of the tick labels
#xtick.labelsize      : medium # fontsize of the tick labels
#xtick.direction      : out     # direction: in or out
```

## 7. adding a legend

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(1., 8., 30)
ax.plot(x, x ** 1.5, 'ro', label='density')
ax.plot(x, 20/x, 'bx', label='temperature')
ax.legend()
plt.rc('legend', fontsize='small')
```

## 8. adding a colorbar

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
image = np.random.poisson(10., (100, 80))
i = ax.imshow(image, interpolation='nearest')
fig.colorbar(i)  # note that colorbar is a method of the figure, not the axes
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.6,0.8])
image = np.random.poisson(10., (100, 80))
i = ax.imshow(image, interpolation='nearest')
colorbar_ax = fig.add_axes([0.7, 0.1, 0.05, 0.8])
fig.colorbar(i, cax=colorbar_ax)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.6,0.8])
image = np.random.poisson(10., (100, 80))
i = ax.imshow(image, aspect='auto', interpolation='nearest')
colorbar_ax = fig.add_axes([0.7, 0.1, 0.05, 0.8])
fig.colorbar(i, cax=colorbar_ax)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.6, 0.8])
x = np.random.random(400)
y = np.random.random(400)
c = np.random.poisson(10., 400)
s = ax.scatter(x, y, c=c, edgecolor='none')
ax.set_xlim(0., 1.)
ax.set_ylim(0., 1.)
colorbar_ax = fig.add_axes([0.7, 0.1, 0.05, 0.8])
fig.colorbar(s, cax=colorbar_ax)
```

```python
# exercise 1

fig=plt.figure(figsize=(8,6)) # initialize figure and axes
ax=fig.add_axes([0.1,0.1,0.8,0.8])

spectral_id = [1, 2, 3, 4, 5, 6, 7] # Define spectral types
spectral_types = ['O', 'B', 'A', 'F', 'G', 'K', 'M']

ax.plot(spectral_id, [4, 3, 2, 3, 4, 5, 4], 'ro') # Plot the data

ax.set_xlim(0.5, 7.5) # Set the limits
ax.set_ylim(0., 10.)

ax.set_xticks(spectral_id) # Set the custom ticks on the x-axis
ax.set_xticklabels(spectral_types)

ax.set_xlabel('SPECTRAL TYPE') # set the axis labels
ax.set_ylabel('NUMBER OF SOURCES')
```
